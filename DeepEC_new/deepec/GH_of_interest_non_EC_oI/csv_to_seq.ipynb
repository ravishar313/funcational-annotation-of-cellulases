```python
acc_vs_seq=pd.read_csv('Sequences.csv') #read the csv file containing the length of each sequence as well, obtained from Get_EC_GH.ipynb
#acc_vs_seq.rename( columns={'Unnamed: 0':'ID'}, inplace=True ) #rename the unnamed column as ID
acc_vs_seq
```

```python
acc_vs_seq=acc_vs_seq[acc_vs_seq.Length <1000] #removing sequences more than 1000 and less than 60 amino acids long
acc_vs_seq=acc_vs_seq[acc_vs_seq.Length >60]
acc_vs_seq.reset_index(inplace = True, drop = True) #resetting the index, as after dropping there are missing index values
acc_vs_seq
```

```python
#saving the sequences to multifasta after length trimmed
xx=(acc_vs_seq.ID.count())
with open('GH_oI_EC_oNI.fasta','w',encoding='utf-8') as otp:
    for i in range(xx):
        acc_id=acc_vs_seq.iloc[i,0]+'&'+acc_vs_seq.iloc[i,1]+'&'+acc_vs_seq.iloc[i,3]
        seq=acc_vs_seq.iloc[i,4]
        otp.write('>'+acc_id+'\n'+seq+'\n')
print('multifasta file with no length outliers is successfully saved')
```
