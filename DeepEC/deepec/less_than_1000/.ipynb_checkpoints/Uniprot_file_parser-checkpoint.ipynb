```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_table('post_feb_2018.txt')
df
```

```python
EC_df=df[~pd.isna(df['EC number'])]
```

```python
EC_df.reset_index(inplace=True)
EC_df
```

```python
EC_df=EC_df.drop('index', axis='columns')
EC_df
```

```python
not_EC_df=df[pd.isna(df['EC number'])]
```

```python
not_EC_df.reset_index(inplace=True)
```

```python
not_EC_df=not_EC_df.drop('index', axis='columns')
not_EC_df
```

```python
EC_df.to_csv('EC_df.csv')
```

```python
not_EC_df.to_csv('not_EC_df.csv')
```

```python
#saving fasta file for EC numbers
with open('EC.fasta', 'w') as file:
    for i in range(len(EC_df.Entry)):
        file.write('>'+EC_df['Entry'][i]+'@'+EC_df['Entry Name'][i]+'@'+EC_df['EC number'][i]+'\n')
        file.write(EC_df['Sequence'][i]+'\n')
```

```python
#saving fasta file for non-EC numbers
with open('not_EC.fasta', 'w') as file:
    for i in range(len(not_EC_df.Entry)):
        file.write('>'+not_EC_df['Entry'][i]+'@'+not_EC_df['Entry Name'][i]+'\n')
        file.write(not_EC_df['Sequence'][i]+'\n')
```
